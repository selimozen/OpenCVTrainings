```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
orion = cv2.imread("orion.jpg")
orion = cv2.cvtColor(orion, cv2.COLOR_BGR2RGB)
```

```python
cateye = cv2.imread("cateye.jpg")
cateye = cv2.cvtColor(cateye, cv2.COLOR_BGR2RGB)
```

```python
rozet = cv2.imread("rozet.jpg")
rozet = cv2.cvtColor(rozet, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(20,15))
plt.subplot(1,3,1)
plt.imshow(orion)
plt.subplot(1,3,2)
plt.imshow(cateye)
plt.subplot(1,3,3)
plt.imshow(rozet)
```

```python
#channel fotorafmzn hangi kanal tonlarnda renk dalmn incelememiz iin tanmlanr. Gri tonlar iin 0, krmz iin 2
```

```python
hist_values = cv2.calcHist([orion], channels = [2], mask = None, histSize = [256], ranges = [0,256])
```

```python
plt.plot(hist_values)
```

```python
plt.figure(figsize=(20,15))
plt.subplot(3,2,1)
color = ('b', 'g', 'r')
for i,col in enumerate(color):
    histr = cv2.calcHist([orion],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.subplot(3,2,2)
hist_values1 = cv2.calcHist([orion], channels = [2], mask = None, histSize = [256], ranges = [0,256])
plt.plot(hist_values1)
plt.subplot(3,2,3)
for i,col in enumerate(color):
    histr = cv2.calcHist([cateye],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.subplot(3,2,4)
hist_values2 = cv2.calcHist([cateye], channels = [2], mask = None, histSize = [256], ranges = [0,256])
plt.plot(hist_values2)
plt.subplot(3,2,5)
for i,col in enumerate(color):
    histr = cv2.calcHist([rozet],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.subplot(3,2,6)
hist_values3 = cv2.calcHist([rozet], channels = [2], mask = None, histSize = [256], ranges = [0,256])
plt.plot(hist_values3)
plt.show()
```
